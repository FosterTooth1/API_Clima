```python
import pandas as pd
import numpy as np
```

```python
# Cargar el CSV
df_naves_industriales = pd.read_csv('../data/raw/CSV_Sucio_Naves.csv', delimiter=',')
```

```python
# Inspeccionar el CSV
print(df_naves_industriales.info())
```

```python
# Dropear columnas innecesarias
df_naves_industriales = df_naves_industriales.drop(columns=['N','MUNICIPIO','DIRECCION','ADMINISTRACION'])

# Renombrar columnas para facilitar el acceso
df_naves_industriales = df_naves_industriales.rename(columns={
    'LATITUD': 'latitud',
    'LONGITUD': 'longitud',
    'PARQUE INDUSTRIAL': 'nombre'
})

# Ordenar por orden alfabético de nombre
df_naves_industriales = df_naves_industriales.sort_values(by='nombre')

# Convertir en tipo oración la columna nombre (Ejemplo: "parque industrial la granja" -> "Parque Industrial La Granja")
df_naves_industriales['nombre'] = df_naves_industriales['nombre'].str.title()

# Rsetear el índice
df_naves_industriales = df_naves_industriales.reset_index(drop=True)

print(df_naves_industriales.info())
```

```python
# Guardar el DataFrame limpio a un nuevo CSV
df_naves_industriales.to_csv('../data/processed/Naves_Industriales.csv', index=False)
```
