```python
import pandas as pd
import numpy as np
```

```python
# Cargar el CSV
df_municipios = pd.read_csv('../data/raw/Municipios_Sucios.csv', delimiter=',')
```

```python
# Inspeccionar el CSV
print(df_municipios.info())
```

```python
# Dropear columnas innecesarias
df_municipios = df_municipios.drop(columns=['clave_entidad','clave_municipio','poblacion'])

# Conservar solo los registros con el valor 'México' en la columna 'nombre_entidad'
df_municipios = df_municipios[df_municipios['entidad'] == 'México']

# Dropear la columna 'entidad' ya que ahora todos los registros son de 'México'
df_municipios = df_municipios.drop(columns=['entidad'])

# Renombrar columnas para mayor claridad
df_municipios = df_municipios.rename(columns={'municipio': 'nombre', 'latitud': 'lat', 'longitud': 'lon'})

# Mover la columna lon al final del DataFrame
lon = df_municipios.pop('lon')
df_municipios['lon'] = lon

# Resetear el índice del DataFrame
df_municipios = df_municipios.reset_index(drop=True)


# Inspeccionar el DataFrame resultante
print(df_municipios.info())
```

```python
# Guardar el DataFrame limpio a un nuevo CSV
df_municipios.to_csv('../data/processed/Municipios.csv', index=False)
```
